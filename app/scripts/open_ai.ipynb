```python
#!/usr/bin/env python3
"""
Read text1.txt and keep every other line (1,3,5,...), then print the result.
"""

from pathlib import Path

def main(path="mock_scripts/text1.txt"):
    p = Path(path)
    if not p.exists():
        raise FileNotFoundError(f"Missing file: {p.resolve()}")

    # Collect into a list (efficient), then join into a single string
    kept = []
    with p.open("r", encoding="utf-8", errors="replace") as f:
        for i, line in enumerate(f, start=1):  # 1-indexed
            if i % 2 != 1:                     # keep lines 1,3,5,...
                kept.append(line)

    result = "".join(kept)  # one big string with original newlines preserved
    print(result)

if __name__ == "__main__":
    main()
```

```python
import re, math
from collections import defaultdict
from typing import Dict, Tuple, List

HASHTAG_RE = re.compile(r"#\w+", flags=re.UNICODE)

def _extract_hashtags(text: str) -> List[str]:
    return [h.lower() for h in HASHTAG_RE.findall(text or "")]

def _rank(d: Dict[str, Tuple[int, int]]) -> List[Tuple[str, int, int, float]]:
    """
    d[tag] = (count, total_views)
    returns list sorted by score where score = count * log10(total_views+1)
    """
    out = []
    for tag, (cnt, views) in d.items():
        score = cnt * math.log10(views + 1)
        out.append((tag, cnt, views, score))
    return sorted(out, key=lambda x: x[3], reverse=True)

def top_hashtags_for_query(query: str, max_results: int = 50, order: str = "viewCount", *, max_pages: int = 1):
    """
    Finds recent/top videos for a query and returns ranked hashtags.
    order: 'viewCount' | 'relevance' | 'date'
    """
    yt = get_youtube_service()
    video_ids = []
    page_token = None
    pages = 0

    while pages < max_pages:
        resp = yt.search().list(
            q=query, type="video", part="id",
            maxResults=min(50, max_results),
            order=order, pageToken=page_token,
            safeSearch="none"
        ).execute()
        video_ids.extend([item["id"]["videoId"] for item in resp.get("items", [])])
        page_token = resp.get("nextPageToken")
        pages += 1
        if not page_token or len(video_ids) >= max_results:
            break

    tags: Dict[str, Tuple[int, int]] = defaultdict(lambda: [0, 0])  # count, views

    for i in range(0, len(video_ids), 50):
        chunk = video_ids[i:i+50]
        v = yt.videos().list(part="snippet,statistics", id=",".join(chunk)).execute()
        for it in v.get("items", []):
            snip = it["snippet"]
            stats = it.get("statistics", {})
            views = int(stats.get("viewCount", 0))
            title = snip.get("title", "")
            desc = snip.get("description", "")
            hs = set(_extract_hashtags(title) + _extract_hashtags(desc))
            for h in hs:
                tags[h][0] += 1
                tags[h][1] += views

    ranked = _rank(tags)
    print(f"\nTop hashtags for query: {query}")
    print(f"{'hashtag':<24} {'count':>5} {'views':>12} {'score':>8}")
    for tag, cnt, views, score in ranked[:25]:
        print(f"{tag:<24} {cnt:>5} {views:>12} {score:>8.2f}")
    return ranked

def top_hashtags_for_channel(channel_id: str, max_results: int = 100):
    """
    Scans a channel’s latest uploads for hashtags.
    """
    yt = get_youtube_service()
    # Get Uploads playlist
    ch = yt.channels().list(part="contentDetails", id=channel_id).execute()
    uploads_pl = ch["items"][0]["contentDetails"]["relatedPlaylists"]["uploads"]

    # Pull recent video IDs from uploads playlist
    video_ids = []
    page_token = None
    while True:
        pl = yt.playlistItems().list(
            part="contentDetails", playlistId=uploads_pl,
            maxResults=50, pageToken=page_token
        ).execute()
        video_ids.extend([it["contentDetails"]["videoId"] for it in pl.get("items", [])])
        page_token = pl.get("nextPageToken")
        if not page_token or len(video_ids) >= max_results:
            break

    tags: Dict[str, Tuple[int, int]] = defaultdict(lambda: [0, 0])

    for i in range(0, len(video_ids), 50):
        chunk = video_ids[i:i+50]
        v = yt.videos().list(part="snippet,statistics", id=",".join(chunk)).execute()
        for it in v.get("items", []):
            snip = it["snippet"]; stats = it.get("statistics", {})
            views = int(stats.get("viewCount", 0))
            hs = set(_extract_hashtags(snip.get("title","")) + _extract_hashtags(snip.get("description","")))
            for h in hs:
                tags[h][0] += 1
                tags[h][1] += views

    ranked = _rank(tags)
    print(f"\nTop hashtags for channel: {channel_id}")
    print(f"{'hashtag':<24} {'count':>5} {'views':>12} {'score':>8}")
    for tag, cnt, views, score in ranked[:25]:
        print(f"{tag:<24} {cnt:>5} {views:>12} {score:>8.2f}")
    return ranked
```
